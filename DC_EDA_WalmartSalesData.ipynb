```python
## Loading the Walmart Sales Dataset
import pandas as pd

# Read CSV into a DataFrame (do NOT pass a filename string to DataFrame constructor)
df = pd.read_csv('Walmart_sales_data.csv')

# Quick preview
df.head()
```

```python
## Getting basic info and statistics about the data

df.info()
df.describe()
```

```python
## Getting the null values

df.isnull().sum()
```

```python
## Filling the null values for numerical and categorical columns

df['Unit_Price'] = df['Unit_Price'].fillna(df['Unit_Price'].mean())
df['Customer_Segment'] = df['Customer_Segment'].fillna(df['Customer_Segment'].mode()[0])
```

```python

## Check if we still have missing values
df.isnull().sum()
```
